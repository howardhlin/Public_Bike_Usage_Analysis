```python
import pandas as pd
import numpy as np
```

```python
Daily_usage = pd.read_csv('/Users/howardlin/Desktop/DEPA data/dock turnover rate/daily_usage_by_region.csv', skiprows=0)
Traffic_speed = pd.read_csv('/Users/howardlin/Desktop/DEPA data/dock turnover rate/Traffic_Speed.csv', skiprows=0)
```

```python
region = pd.read_csv('/Users/howardlin/Desktop/DEPA data/region_cleaned.csv', skiprows=0)
```

```python
Daily_usage=region.merge(Daily_usage, left_on='REGION_NAME', right_on='region_name', how='inner')
```
