```python
import pandas as pd
import numpy as np
```

```python
startstation = pd.read_csv('/Users/howardlin/Desktop/DEPA data/Bike-out vs bike-in/tableau_startstation_vs_trip1.csv', skiprows=0)
endstation = pd.read_csv('/Users/howardlin/Desktop/DEPA data/Bike-out vs bike-in/tableau_endstation_vs_trip1.csv', skiprows=0)
```

```python
endstation = endstation.fillna(value={'Morning': 0, 'Afternoon': 0, 'Evening': 0,'Night': 0})
```

```python
startstation = startstation.fillna(value={'Morning': 0, 'Afternoon': 0, 'Evening': 0,'Night': 0})
```

```python
startmorning = startstation.iloc[:,0:4]
startafternoon = startstation.iloc[:,[0,1,2,4]]
startevening = startstation.iloc[:,[0,1,2,5]]
startnight = startstation.iloc[:,[0,1,2,6]]
endmorning = endstation.iloc[:,0:4]
endafternoon = endstation.iloc[:,[0,1,2,4]]
endevening = endstation.iloc[:,[0,1,2,5]]
endnight = endstation.iloc[:,[0,1,2,6]]
```

```python
morningmerge=startmorning.merge(endmorning, left_on='Start Station Key', right_on='End Station Key', how='inner')
afternoonmerge=startafternoon.merge(endafternoon, left_on='Start Station Key', right_on='End Station Key', how='inner')
eveningmerge=startevening.merge(endevening, left_on='Start Station Key', right_on='End Station Key', how='inner')
nightmerge=startnight.merge(endnight, left_on='Start Station Key', right_on='End Station Key', how='inner')
```

```python
nightmerge
```

```python
nightmerge[nightmerge['Start Station Key']-nightmerge['End Station Key']==0]
```

```python
morningmerge['diff'] = (morningmerge['Morning_x']-morningmerge['Morning_y'])
afternoonmerge['diff'] = (afternoonmerge['Afternoon_x']-afternoonmerge['Afternoon_y'])
eveningmerge['diff'] = (eveningmerge['Evening_x']-eveningmerge['Evening_y'])
nightmerge['diff'] = (nightmerge['Night_x']-nightmerge['Night_y'])
```

```python
morning1 = morningmerge.drop(columns=['End Station Key','Morning_x','Station Name (Dim Station1)','Station Geometry (Dim Station1)','Morning_y'])
afternoon1 = afternoonmerge.drop(columns=['End Station Key','Afternoon_x','Station Name (Dim Station1)','Station Geometry (Dim Station1)','Afternoon_y'])
evening1 = eveningmerge.drop(columns=['End Station Key','Evening_x','Station Name (Dim Station1)','Station Geometry (Dim Station1)','Evening_y'])
night1 = nightmerge.drop(columns=['End Station Key','Night_x','Station Name (Dim Station1)','Station Geometry (Dim Station1)','Night_y'])
```

```python
morning1["bike-out more than bike-in"]=(morning1['diff']>0)
afternoon1["bike-out more than bike-in"]=(afternoon1['diff']>0)
evening1["bike-out more than bike-in"]=(evening1['diff']>0)
night1["bike-out more than bike-in"]=(night1['diff']>0)
```

```python
morning1.to_csv('/Users/howardlin/Desktop/DEPA data/Bike-out vs bike-in/morning.csv',index=False)
afternoon1.to_csv('/Users/howardlin/Desktop/DEPA data/Bike-out vs bike-in/afternoon.csv',index=False)
evening1.to_csv('/Users/howardlin/Desktop/DEPA data/Bike-out vs bike-in/evening.csv',index=False)
night1.to_csv('/Users/howardlin/Desktop/DEPA data/Bike-out vs bike-in/night.csv',index=False)
```
