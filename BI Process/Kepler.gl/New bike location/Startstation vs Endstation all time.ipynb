```python
import pandas as pd
import numpy as np
```

```python
startstation = pd.read_csv('/Users/howardlin/Desktop/DEPA data/Bike-out vs bike-in/tableau_startstation_vs_trip1.csv', skiprows=0)
endstation = pd.read_csv('/Users/howardlin/Desktop/DEPA data/Bike-out vs bike-in/tableau_endstation_vs_trip1.csv', skiprows=0)
```

```python
startstation['Total']=startstation['Morning']+startstation['Afternoon']+startstation['Evening']+startstation['Night']
endstation['Total']=endstation['Morning']+endstation['Afternoon']+endstation['Evening']+endstation['Night']
```

```python
startstation
```

```python
endstation = endstation.fillna(value={'Morning': 0, 'Afternoon': 0, 'Evening': 0,'Night': 0})
```

```python
startstation = startstation.fillna(value={'Morning': 0, 'Afternoon': 0, 'Evening': 0,'Night': 0})
```

```python
startstation = startstation.drop(columns=['Morning','Afternoon','Evening','Night'])
endstation = endstation.drop(columns=['Morning','Afternoon','Evening','Night'])
```

```python
merges=startstation.merge(endstation, left_on='Start Station Key', right_on='End Station Key', how='inner')
```

```python
merges
```

```python
merges[merges['Start Station Key']-merges['End Station Key']==0]
```

```python
merges['diff'] = (merges['Total_x']-merges['Total_y'])
```

```python
merge1 = merges.drop(columns=['End Station Key','Station Name (Dim Station1)','Station Geometry (Dim Station1)'])
```

```python
merge1["bike-out more than bike-in"]=(merge1['diff']>0)
```

```python
merge1.to_csv('/Users/howardlin/Desktop/DEPA data/Bike-out vs bike-in/total.csv',index=False)
```

```python
ride_perdock = pd.read_csv('/Users/howardlin/Desktop/DEPA data/dock turnover rate/Ridership_per_Dock.csv', skiprows=0)
```

```python
merge2=merge1.merge(ride_perdock, left_on='Start Station Key', right_on='station_key', how='inner')
```

```python
merge2=merge2.drop(columns=['station_key','station_name','station_geometry'])
```
